```python
import pandas as pd
```

```python
from yelp_preprocessing import YelpPreprocessor
from yelp_eda import YelpEDA
```

```python
# Initialize Preprocessor with file path
preprocessor = YelpPreprocessor(
    data_dir='data/',
    photo_dir='data/photos'
)

# Set use_sample=True for faster prototyping with a subset of data
preprocessor.run_full_pipeline(use_sample=True, sample_frac=0.1)
```

```python
# Initialize EDA module
eda = YelpEDA(preprocessor, output_dir='eda_outputs')

# Run full EDA
eda.run_full_eda()
```
